```python
import pandas as pd

# Créer un DataFrame simple
data = {
    "francais": ["maison", "école", "nourriture", "eau", "travail"],
    "comorien": ["nyumba", "skuli", "chakula", "maji", "kazi"]
}

df = pd.DataFrame(data)

# Sauvegarder si besoin
# df.to_csv('mots_francais_comorien.csv', index=False)

print(df)
```

```python
from sentence_transformers import SentenceTransformer

# Charger le modèle multilingue
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

# Générer les embeddings
embeddings_francais = model.encode(df['francais'].tolist())
embeddings_comorien = model.encode(df['comorien'].tolist())
```

```python
pip install torch
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Transformer les données en tenseurs PyTorch
X = torch.tensor(embeddings_comorien, dtype=torch.float32)
y = torch.tensor(embeddings_francais, dtype=torch.float32)

# Mini modèle
class EmbeddingAligner(nn.Module):
    def __init__(self, dimension):
        super(EmbeddingAligner, self).__init__()
        self.fc = nn.Linear(dimension, dimension)  # simple couche linéaire
        
    def forward(self, x):
        return self.fc(x)

# Paramètres
embedding_dim = X.shape[1]
model = EmbeddingAligner(embedding_dim)
optimizer = optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

# Entraînement rapide
for epoch in range(200):
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()
    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```
